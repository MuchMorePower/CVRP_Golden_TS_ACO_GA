```python
import matplotlib.pyplot as plt
import numpy as np


plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False


algorithms = ['TS', 'ACO', 'PMA', 'HMA+LS', 'HMA']
datasets = ['路径密集型', '路径稀疏型']
metrics = ['平均Gap', '平均计算时间']


data = {
    '路径密集型': {
        '平均Gap': [15.2, 13.8, 96.3, 15.6, 15.6],
        '平均计算时间': [191.7, 1113.6, 1603.5, 795.7, 24.4]
    },
    '路径稀疏型': {
        '平均Gap': [19.3, 16.6, 147.0, 16.5, 17.2],
        '平均计算时间': [576.5, 969.5, 126.8, 89.9, 15.2]
    }
}


fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=False, dpi=1000)


bar_width = 0.15
index = np.arange(len(datasets)) 


colors = ['#BAE3DC', '#DAD7EA', '#F6B3AC', '#B3D1E7', '#F7D7B0']


for algo_idx, algorithm in enumerate(algorithms):
    values_optimal = [data[d][metrics[0]][algo_idx] for d in datasets]
    values_time = [data[d][metrics[1]][algo_idx] for d in datasets]

    
    axes[0].bar(index + algo_idx * bar_width,
                values_optimal,
                width=bar_width,
                label=algorithm,
                color=colors[algo_idx],
                )

    
    axes[1].bar(index + algo_idx * bar_width,
                values_time,
                width=bar_width,
                label=algorithm,
                color=colors[algo_idx],
                )


axes[0].set_title('平均Gap对比', fontsize=18)
axes[0].set_xlabel('数据集类型', fontsize=18)
axes[0].set_ylabel('最优解值', fontsize=18)
axes[0].set_xticks(index + bar_width * (len(algorithms) / 2))
axes[0].set_xticklabels(datasets)
axes[0].legend(title='算法')
axes[0].grid(True, linestyle='--', alpha=0.5)


axes[1].set_title('平均计算时间对比', fontsize=18)
axes[1].set_xlabel('数据集类型', fontsize=18)
axes[1].set_ylabel('计算时间 (s)', fontsize=18)
axes[1].set_xticks(index + bar_width * (len(algorithms) / 2))
axes[1].set_xticklabels(datasets)
axes[1].grid(True, linestyle='--', alpha=0.5)

# 添加数值标签
for ax in axes:
    for bars in ax.containers:
        for bar in bars:
            height = bar.get_height()
            ax.annotate(f'{height}',
                        xy=(bar.get_x() + bar.get_width() / 2, height),
                        xytext=(0, 3),
                        textcoords="offset points",
                        ha='center', va='bottom', fontsize=12)

plt.tight_layout()

plt.show()
```
